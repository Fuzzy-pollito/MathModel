```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python

def load_images(folder):
    images = []
    labels = []
    imgforhistogram = []
    for fname in os.listdir(folder):
        img = Image.open(os.path.join(folder, fname)).convert('L')  # grayscale
        img_array = np.array(img).flatten() / 255.0
        imgforhistogram.append(np.median(img_array))
    return imgforhistogram

img_h=load_images('data - Copy/Train')
```

```python
plt.hist(img_h, bins=np.linspace(0,1,50))
```

```python
import numpy as np
from skimage.filters import threshold_otsu

# Suppose means is a 1D array of all image means:
means = np.array([np.mean(Image.open(os.path.join('data - Copy/Train', f)).convert('L'))
                  for f in os.listdir('data - Copy/Train')]) / 255.0

# Compute Otsu threshold on the means
otsu_thresh = threshold_otsu(means)
print("Otsu threshold:", otsu_thresh)

# Inversion
```

```python
import matplotlib.pyplot as plt

# 1. Load images without flipping
means = []
for fname in os.listdir('data - Copy/Train'):
    img = Image.open(os.path.join('data - Copy/Train', fname)).convert('L')
    img_array = np.array(img).flatten() / 255.0
    means.append(np.median(img_array))

# 2. Plot histogram
plt.hist(means, bins=50, color='skyblue')
plt.xlabel('Mean Intensity')
plt.ylabel('Number of Images')
plt.title('Histogram of Image Mean Intensities')
plt.show()
```
